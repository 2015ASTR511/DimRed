```python
"""
Code to run PCA on chunks of APOGEE spectra in bins of 100K width in Teff.
  
Authors: Grace Telford, Diana Windemuth, Jake Lustig-Yager
Date: 2/18/2015
For Dimensionality Reduction Project, ASTR 511 

Updated 2/22/2015 by Diana Windemuth
Updated 2/25/2015 by Grace Telford
Updated 2/28/2015 by Jake Lustig-Yager
Updated 3/2/2015 by Grace Telford
"""

import numpy as np
import pyfits #maybe should use astropy.io instead?? 
from sklearn.decomposition import PCA
import matplotlib as mpl
mpl.rc('xtick', labelsize=14)     
mpl.rc('ytick', labelsize=14)  
mpl.rc('axes', labelsize=16)
mpl.rc('legend', fontsize=16)
#mpl.rc('lines', linewidth=2)
mpl.rc('font', size=16, family='serif', style='normal', variant='normal', stretch='normal', weight='normal')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""
Read in allStar table, mask out rows where LOCATION_ID=1.
Read in one chunk of spectra at a time.
Read one apStar file to get wavelength grid info.
Apply mask to select parameters for stars in each chunk.
"""

def readSpecChunk(specname='spectra4000.npy', flagname='flags4000.npy', maskname='mask4000.npy'):
    spectra = np.load('/astro/store/scratch/tmp/DimRed/dataChunks/'+specname)
    flags = np.load('/astro/store/scratch/tmp/DimRed/dataChunks/'+flagname)
    mask = np.load('/astro/store/scratch/tmp/DimRed/dataChunks/'+maskname)
    
    return spectra, flags, mask


def getWavelengths():
    # pick one file; same wavelength grid for all
    file = pyfits.open('/astro/store/scratch/tmp/DimRed/apStar/apStar-r5-2M00000211+6327470.fits')
    
    start = file[0].header['CRVAL1']
    delta = file[0].header['CDELT1']
    N_wave = file[0].header['NWAVE']
    end = start + N_wave * delta
    
    wavelength = np.logspace(start, end, N_wave)
    
    file.close()
    
    return wavelength


def readParams(dataname='allStar-v603.fits'):
    allStar = pyfits.open('/astro/store/scratch/tmp/DimRed/allStar-v603.fits')

    fields = allStar[1].data['LOCATION_ID']
    field_mask = fields != 1

    Teff = allStar[1].data['TEFF'][field_mask]
    logg = allStar[1].data['LOGG'][field_mask]
    FeH = allStar[1].data['PARAM_M_H'][field_mask]
    alphaFe = allStar[1].data['PARAM_ALPHA_M'][field_mask]

    allStar.close()
    
    return Teff, logg, FeH, alphaFe

    
def applyMask(mask, Teff, logg, FeH, alphaFe):
    
    Teff = Teff[mask]
    logg = logg[mask]
    FeH = FeH[mask]
    alphaFe = alphaFe[mask]
    
    # check number of disk/halo stars
    condHalo = (FeH > -2.5) * (FeH < -1.1)
    condDisk = (FeH > -0.9) * (FeH < 0.6) 
    
    print 'Number of halo stars: ', np.sum(condHalo)
    print 'Number of disk stars: ', np.sum(condDisk)
    
    return Teff, logg, FeH, alphaFe
```

```python
def runPCA(wavelength, spectra, n_components=10, plot=True, figtitle=None):

    pca = PCA(n_components=n_components)

    # mask out chip gaps
    chipgaps = (wavelength < 15170.) + ((wavelength > 15800.)*(wavelength < 15860.)) \
                + ((wavelength > 16430.)*(wavelength < 16480.)) + (wavelength > 16950.)
    #errormask = (spectra[1, :, :] == 1e10)
    #finalmask = (np.sum(errormask, axis=0) > spectra.shape[1]*.01)
    finalmask = ~chipgaps
    spectra = spectra[:, :, finalmask]
    wavelength = wavelength[finalmask]

    # try dividing by mean instead of subtracting; otherwise still have HUGE values in mean spectrum
    #spectra[0, :, :] -= np.mean(spectra[0, :, :], axis=1)[:, np.newaxis]
    norm_spectra = spectra[0,:,:] / np.mean(spectra[0,:,:], axis=1).reshape(2888,1)
    
    pca.fit(norm_spectra)
    print 'Variances of recovered PCs: ', pca.explained_variance_ratio_

    if plot:
        plt.figure(figsize=(12,8))
        plt.suptitle(figtitle)
        
        plt.subplot(3,2,1)
        plt.ylabel('Mean')
        plt.xlim([15000, 17000])
        plt.plot(wavelength, pca.mean_)

        for ii in range(5):
            plt.subplot(3,2,ii+2)
            plt.ylabel('PC '+str(ii+1))
            plt.xlim([15000, 17000])
            plt.text(15200, np.min(pca.components_[ii]), 'Variance: %g' % pca.explained_variance_ratio_[ii],fontsize=12)
            plt.plot(wavelength, pca.components_[ii])
        plt.tight_layout()

        # make scree plot
        plt.figure(figsize=(8,6))
        plt.plot(np.arange(1,n_components+1), pca.explained_variance_ratio_)
        plt.title(figtitle)
        plt.xlabel('PC')
        plt.ylabel('Variance')
        
        return pca.transform(norm_spectra)#pca.transform(spectra[0,:,:]) 
```

```python
def plotTransform(lowDim, metallicities, Teff, logg, zoom=False):            
        # plot data transformed to space of first two PCs color coded by [Fe/H], Teff, and logg

        transform = lowDim #/10000.
        
        plt.figure(figsize=(15,8))
        
        plt.subplot(2,3,1)
        plt.scatter(transform[:,0], transform[:,1], c=metallicities, s=30, lw=0)
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.colorbar(label='[Fe/H]')
        if zoom:
            plt.xlim([-2,5])
            plt.ylim([-1.5,0.5])

        plt.subplot(2,3,2)
        plt.scatter(transform[:,0], transform[:,1], c=Teff, s=30, lw=0)
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.colorbar(label='Teff')
        if zoom:
            plt.xlim([-2,5])
            plt.ylim([-1.5,0.5])
        
        plt.subplot(2,3,3)
        plt.scatter(transform[:,0], transform[:,1], c=logg, s=30, lw=0)
        plt.xlabel('PC1')
        plt.ylabel('PC2')
        plt.colorbar(label='log(g)')
        if zoom:
            plt.xlim([-2,5])
            plt.ylim([-1.5,0.5])
        
        # PCs 2 and 3
        
        plt.subplot(2,3,4)
        plt.scatter(transform[:,2], transform[:,1], c=metallicities, s=30, lw=0)
        plt.xlabel('PC3')
        plt.ylabel('PC2')
        plt.colorbar(label='[Fe/H]')
        if zoom:
            plt.xlim([-1,1.5])
            plt.ylim([-1.5,0.5])

        plt.subplot(2,3,5)
        plt.scatter(transform[:,2], transform[:,1], c=Teff, s=30, lw=0)
        plt.xlabel('PC3')
        plt.ylabel('PC2')
        plt.colorbar(label='Teff')
        if zoom:
            plt.xlim([-1,1.5])
            plt.ylim([-1.5,0.5])
            
        plt.subplot(2,3,6)
        plt.scatter(transform[:,2], transform[:,1], c=logg, s=30, lw=0)
        plt.xlabel('PC3')
        plt.ylabel('PC2')
        plt.colorbar(label='log(g)')
        if zoom:
            plt.xlim([-1,1.5])
            plt.ylim([-1.5,0.5])        
        plt.tight_layout()
```

```python
# Read in data for first chunk in Teff
spectra, flags, mask = readSpecChunk()
wavelength = getWavelengths()
Teff, logg, FeH, alphaFe = readParams()
Teff, logg, FeH, alphaFe = applyMask(mask, Teff, logg, FeH, alphaFe)

# Apply PCA
lowDim = runPCA(wavelength, spectra, figtitle=r'$4000K < T_{eff} < 4100K$')
```

```python
# plot transformed data color coded by various parameters
plotTransform(lowDim, FeH, Teff, logg)
```

## 2D Color Plots

```python
"""
These functions are used to create 2d color plots.

>>> make_color_plot(alpha,metallicities,num=1)

Adapted from code by Jake VanderPlas <vanderplas@astro.washington.edu>
rom astroML: see http://astroML.github.com
"""
#import numpy as np
#from matplotlib import pyplot as plt
#from astroML.datasets import fetch_moving_objects
from astroML.plotting.tools import devectorize_axes
#import matplotlib as mpl
#import matplotlib.pyplot as plt
from matplotlib import gridspec
#from matplotlib import rc
#from astroML.plotting import setup_text_plots
#setup_text_plots(fontsize=18, usetex=True)
#rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
#mpl.rcParams['font.size'] = 20.0

def compute_color(q1, q2, a_crit=0.1):
    """
    Compute the scatter-plot color using code adapted from
    TCL source used in Parker 2008.
    
    mag_a --> alpha --> q1
    mag_i - mag_z --> metallicities --> q2
    """
    # define the base color scalings
    R = np.ones_like(q1)
    G = 1.0 * 10 ** (8 * (q1 - 0.15))
    B = 1.0 * 10 ** (2 * (q2 - 1.25))

    # enhance green beyond the a_crit cutoff
    #i = np.where(q1 < a_crit)
    #G[i] += 10000 * (10 ** (-0.01 * (q1[i] - a_crit)) - 1)

    # normalize color of each point to its maximum component
    RGB = np.vstack([R, G, B])
    RGB /= RGB.max(0)

    # return an array of RGB colors, which is shape (n_points, 3)
    return RGB.T

def plot_physical(x,y,c,gs):   
    """
    Plots physics quantities. This is the 2d color guide plot. 
    """
    ax0 = plt.subplot(gs[0],axisbg='k')
    for spine in ax0.spines.values():
        spine.set_color('w')
    for tick in ax0.xaxis.get_major_ticks() + ax0.yaxis.get_major_ticks():
        for child in tick.get_children():
            child.set_color('w')
    ax0.scatter(x, y, c=c, s=5, lw=0)
    devectorize_axes(ax0, dpi=400)  
    ax0.set_xlabel(r'[$\alpha$/M]',color='w')
    ax0.set_ylabel(r'[Fe/H]',color='w')
    ax0.set_title('')
    
def plot_PCs(which,PCs,c,gs):
    """
    Plots PC(x) vs. PC(x+1)
    """
    xlabel='PC'+str(which+1)
    ylabel='PC'+str(which+2)
    x = PCs[:,which]
    y = PCs[:,which+1]
    
    ax1 = plt.subplot(gs[which+1],axisbg='k')
    for spine in ax1.spines.values():
        spine.set_color('w')
    for tick in ax1.xaxis.get_major_ticks() + ax1.yaxis.get_major_ticks():
        for child in tick.get_children():
            child.set_color('w')
    ax1.scatter(x, y, c=c, s=5, lw=0)
    devectorize_axes(ax1, dpi=400)
    ax1.set_xlabel(xlabel,color='w')
    ax1.set_ylabel(ylabel,color='w')
    ax1.set_title('')     
    
def make_color_plot(param1,param2,PCs=transform,num=4, cuttitle=False):
    """
    Constructs the black background color plots. Specify the number of principle components to
    plot with num. 
    """
    # print how many stars are being plotted
    print("Plotting "+str(len(param1))+" stars")
    
    # establish color scheme
    color = compute_color(param1, param2)

    # specify plot dimensions according to num
    if num == 1:
        fig = plt.figure(figsize=(15,7),facecolor='k')
        gs = gridspec.GridSpec(1,2)
    elif num == 2:
        fig = plt.figure(figsize=(15,5),facecolor='k')
        gs = gridspec.GridSpec(1,3)
    elif num == 3:
        fig = plt.figure(figsize=(13,16),facecolor='k')
        gs = gridspec.GridSpec(3,2)
        
    # plot the 2d color guide plot
    plot_physical(param1,param2,color,gs)
    
    # plot all principle components specified
    for number in range(num):
        plot_PCs(number,PCs,color,gs)
     
    # follow Jake VanderPlas on this one
    text_kwargs = dict(color='w', fontsize=14,
                   transform=plt.gca().transAxes,
                   ha='center', va='bottom')
    
    plt.tight_layout()
    
    if cuttitle:
        plt.suptitle(cuttitle,color='w')
        plt.subplots_adjust(top=.95)
    
    plt.show()
```
